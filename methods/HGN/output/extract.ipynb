#### 合并

```python
import pandas as pd
select_data = pd.read_excel("/home/linjw/GNNs/HGB/NC/benchmark/methods/baseline/output/save03/select_output.xlsx")
dataset = pd.read_csv(f"/home/linjw/GNNs/HGB/NC/benchmark/data/ERMer/Gene_Entity_v2.csv")
dataset = dataset[['id','EcoGene_Accession:String','ENZYME:String','Swiss_Prot:String','label']]
select_data = select_data.iloc[:,1:]
select_data
```

```python
# 计算id的重复次数
id_counts = select_data['id'].value_counts().reset_index()
id_counts.columns = ['id', 'id_count']

# 将重复次数合并到原始DataFrame
select_data = pd.merge(select_data, id_counts, on='id', how='left')
select_data
```

```python
out = dataset[dataset['id'].isin(select_data.id.values)]
# out = pd.merge(out, select_data[['id','id_count']], on='id', how='left')
out
```

```python
needdata = pd.merge(out, select_data[['id','id_count']], on='id', how='left').drop_duplicates("id")
needdata = needdata.sort_values("id_count",ascending=False)
needdata.columns = ['id', 'EcoGene_Accession:String', 'ENZYME:String', 'Swiss_Prot:String','label', 'Cross-Validation Repetitions']
needdata
```

```python
needdata.to_excel("GNN_Predict_TF_01.xlsx",index=False)
```
